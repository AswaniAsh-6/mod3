```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import csv
```

```python
df=pd.read_csv('CarSalePrice.csv')
print(df)
```

```python
df1=df.head(10)
print(df1)
```

```python
a  = df1 ['YearOfRegistration'].tolist()
b=df1['InitialBuyingPrice'].tolist()
plt.plot(a,b)
plt.xlabel('x')
plt.ylabel('y')
plt.title('CarSale')
plt.show()
```

```python
plt.bar(a,b,color='k')
plt.show()
```

```python
bins=[0,25,50,75,100]
plt.hist(b,bins=bins)
plt.show()
```

```python
plt.scatter(a,b,color='k')
plt.show()
```

```python
plt.plot([],[],label='a',linewidth=5)
plt.plot([],[],label='b',linewidth=5)
plt.stackplot(a,b,colors=['r','pink'])
plt.show()
```
