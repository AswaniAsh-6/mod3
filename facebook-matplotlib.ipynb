```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import csv
```

```python
df=pd.read_csv('facebook.csv')
print(df)
```

```python
df1=df.head(10)
print(df1)
```

```python
a  = df1 ['userid'].tolist()
b=df1['age'].tolist()
c=df1['tenure'].tolist()
plt.plot(a,b)
plt.plot(a,c)
plt.xlabel('userid')
plt.ylabel('y')
plt.title('Facebook')
plt.show()
```

```python
b=df1['age'].tolist()
c=df1['tenure'].tolist()
plt.bar(b,c,color='y')
plt.show()
```

```python
bins=[0,25,50,75,100]
plt.hist(b,bins=bins)
plt.show()
```

```python
from matplotlib import style
style.use('dark_background')
plt.scatter(a,b,color='crimson')
plt.show()
```

```python
plt.plot([],[],label='a',linewidth=5)
plt.plot([],[],label='b',linewidth=5)
plt.plot([],[],label='c',linewidth=5)
plt.stackplot(a,b,c,colors=['y','r','pink'])
plt.show()
```

```python
plt.pie(b)
plt.show()
```
