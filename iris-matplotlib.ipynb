```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import csv
```

```python
df=pd.read_csv('iris.csv')
print(df)
```

```python
#line graph
x = []
y = []

with open('iris.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y.append(row[1])
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Iris')
plt.show()
```

```python
#bar graph
plt.bar(x,y,color='y')
plt.show()
```

```python
plt.hist(y,bins=x)
plt.show()
```

```python
from matplotlib import style
style.use('dark_background')
plt.scatter(x,y,color='crimson')
plt.show()
```
