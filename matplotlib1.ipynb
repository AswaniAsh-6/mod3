```python

from matplotlib import pyplot as plt
plt.plot([2,1,3],[4,5,5])
plt.show()
```

```python
x=[5,8,10]
y=[12,16,6]
plt.plot(x,y)
plt.title('info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
#ggplot
from matplotlib import style
style.use('dark_background')
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'y',label='Line one',linewidth=6)
plt.plot(x2,y2,'b',label='Line two',linewidth=5)
plt.title('Epic info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
#bar graph
plt.bar([5,7,9],[12,16,6],label='Line one',color='y')
plt.bar([6,8,10],[6,15,7],label='Line two',color='k')
plt.title('Epic info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
plt.show()
```

```python
#histogram
population_ages=[22, 87, 5, 43, 56,
              73, 55, 54, 11,
              20, 51, 5, 79, 31,
              27]
bins=[0, 25, 50, 75, 100]
plt.hist(population_ages,bins,histtype='bar',rwidth=0.4,color='purple')
plt.title('Epic info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
#scatter
x=[5,8,10,5,2,7,1]
y=[12,16,6,4,8,2,8]
style.use('dark_background')
plt.scatter(x,y,color='crimson')
plt.title('info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
#area plot
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[6,3,4,3,2]
plt.plot([],[],label='Sleeping',linewidth=5)
plt.plot([],[],label='Eating',linewidth=5)
plt.stackplot(days,sleeping,eating,colors=['chocolate','orange'])
plt.title('info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
